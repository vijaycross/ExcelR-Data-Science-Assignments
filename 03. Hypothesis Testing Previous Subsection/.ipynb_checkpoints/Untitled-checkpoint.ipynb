```python
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

```python
df = pd.read_csv(r'Cutlets.csv')
```

```python
for column in df.columns:
    stat, p = stats.shapiro(df[column])
    print(f"{column}: p-value = {p}")
```

```python
plt.figure(figsize=(10, 6))
df.hist(alpha=0.6, bins=15)
plt.title("Cutlet Diameter Distribution")
plt.xlabel("Diameter")
plt.ylabel("Frequency")
plt.legend(df.columns)
plt.show()
```

```python
statistic, p_value = stats.ttest_ind(df["Unit A"], df["Unit B"], equal_var=True)

print("Test statistic:", statistic)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in cutlet diameters between Unit A and Unit B.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in cutlet diameters between Unit A and Unit B.")
```
