# Problem Statement:

1) Extract reviews of any product from ecommerce website like amazon
2) Perform emotion mining

```python
# importing basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup as bs
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# Download NLTK resources
# nltk.download('stopwords')
# nltk.download('punkt')
# nltk.download('wordnet')
```

```python
# Initialize WordNetLemmatizer
wordnet = WordNetLemmatizer()
```

```python
# 2.1 Connecting to Amazon and downloading reviews
oneplus_reviews = []

for i in range(1, 31):
    url = f"https://www.amazon.in/Test-Exclusive-747/product-reviews/B07DJCVTDN/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews{i}"
    response = requests.get(url)
    soup = bs(response.content, "html.parser")
    reviews = soup.findAll("span", attrs={"class", "a-size-base review-text review-text-content"})
    op = [review.text for review in reviews]
    oneplus_reviews.extend(op)
```

```python
# 2.2 Remove duplicates
oneplus_reviews = list(set(oneplus_reviews))
```

```python
import re
# 2.3 Cleaning the text
def clean_text(text):
    text = re.sub("[^A-Za-z]+", " ", text).lower()  # Remove special characters and convert to lowercase
    text = re.sub(r'^https?:\/\/.*[\r\n]*', '', text)  # Remove hyperlinks
    return text

cleaned_reviews = [clean_text(review) for review in oneplus_reviews]
```

```python
# Tokenization and removing stopwords
stop_words = set(stopwords.words('english'))
tokenized_reviews = [word_tokenize(review) for review in cleaned_reviews]
filtered_reviews = [[word for word in tokens if word not in stop_words] for tokens in tokenized_reviews]
```

```python
# 2.4 Create the DataFrame
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([" ".join(review) for review in filtered_reviews])
feature_names = tfidf_vectorizer.get_feature_names_out()
df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
```

```python
# Create a WordCloud
word_list = ' '.join(feature_names)
wordcloud = WordCloud(
    background_color='black',
    width=1800,
    height=1400
).generate(word_list)
```

```python
# Plot the WordCloud
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# 3 Sentiment Analysis
# Load positive and negative words
with open("positive-words.txt", "r") as pw:
    positive_words = [line.strip() for line in pw.readlines() if line.strip()]

with open("negative-words.txt", "r", encoding='latin-1') as nw:
    negative_words = [line.strip() for line in nw.readlines() if line.strip()]
```

```python
# Filter words in positive and negative word lists
positive_reviews = " ".join([word for word in feature_names if word in positive_words])
negative_reviews = " ".join([word for word in feature_names if word in negative_words])
```

```python
# Create WordClouds for positive and negative words
wordcloud_pos = WordCloud(
    background_color='white',
    width=1800,
    height=1400
).generate(positive_reviews)

wordcloud_neg = WordCloud(
    background_color='black',
    width=1800,
    height=1400
).generate(negative_reviews)
```

```python
# Plot the WordClouds for positive and negative words
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(wordcloud_pos)
plt.title("Positive Words")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(wordcloud_neg)
plt.title("Negative Words")
plt.axis("off")

plt.tight_layout()
plt.show()
```
