```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'Q9_a.csv')
```

```python
df.head()
```

```python
df = df.set_index('Index')
```

```python
sns.displot(x='speed', data=df, kde=True);
```

```python
df['speed'].skew()
```

```python
# Tail of the distribution is skewed to the left.
```

```python
df['speed'].kurtosis()
```

```python
# The peakness of the distribution is light.
```

```python
sns.displot(x='dist', data=df, kde=True);
```

```python
df['dist'].skew()
```

```python
# Tail of the distribution is skewed to the Right.
```

```python
df['dist'].kurtosis()
```

```python
# The peakness of the distribution is heavy peak.
```
