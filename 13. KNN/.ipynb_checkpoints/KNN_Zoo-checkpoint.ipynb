# Data Splitting

```python
X = df.drop('type',axis=1)
y = df['type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

# Model Selection

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(scaled_X_train,y_train)
```

```python
knn.get_params()
```

```python
y_pred = knn.predict(scaled_X_test)
```

```python
y_pred
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred)).plot();
```

```python
print(classification_report(y_test, y_pred))
```

# Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
operations = [('scaler',scaler),('knn',knn)]
```

```python
pipe = Pipeline(operations)
```

# Cross Validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values = list(range(1,30))
param_grid = {'knn__n_neighbors':k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe, param_grid, scoring='accuracy', cv=3)
```

```python
full_cv_classifier.fit(X_train,y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
full_cv_classifier.cv_results_
```

```python
pd.DataFrame(full_cv_classifier.cv_results_)['mean_test_score']
```

```python
plt.figure(figsize=[8,6])
plt.plot(pd.DataFrame(full_cv_classifier.cv_results_)['mean_test_score'], marker='o', markersize=4)
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.title('K_value VS Accuracy')
plt.xticks(k_values)
plt.grid()
plt.show();
```

```python
# As shown in the graph, with K=5 we can achive accurary of 90%.
```

# Model Evaluation

```python
y_pred = full_cv_classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred)).plot();
```

```python
print(classification_report(y_test, y_pred))
```
