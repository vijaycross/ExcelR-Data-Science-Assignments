# Book Recommendation for user

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
```

```python
df1.head()
```

```python
# Collaborative Book Recommendation System
# Item-Based Collaborative Filtering
user_data_df = df1.pivot_table(index='userID', columns='title', values='rating').fillna(0)
```

```python
# Calculate Cosine Similarity between Users
user_sim = 1 - pairwise_distances(user_data_df.values, metric='cosine')
user_sim_df = pd.DataFrame(user_sim, index=user_data_df.index, columns=user_data_df.index)
np.fill_diagonal(user_sim, 0)
```

# Model Building

```python
# Recommend data based on Cosine Similarity
def recommend_data(userID):
    similar_users = user_sim_df.sort_values(userID, ascending=False).head(100).index
    recommended_data = set()
    
    for similar_user in similar_users:
        user_data = df1[df1['userID'] == similar_user]['title']
        recommended_data.update(user_data)
    
    user_data_read = df1[df1['userID'] == userID]['title']
    recommendations = recommended_data - set(user_data_read)
    
    return recommendations
```

```python
# Example usage:
recommended_data_for_user_14 = recommend_data(14)
recommended_data_for_user_8 = recommend_data(8)
```

```python
print("Recommended data for User 14:")
print(pd.DataFrame(recommended_data_for_user_14))
```

```python
print("\nRecommended data for User 8:")
print(pd.DataFrame(recommended_data_for_user_8))
```

#  Conditional Recommender System by Books

```python
df1.head()
```

```python
# filtering users who rated more than 10 books

x = df1.groupby('userID').count()['rating'] > 10
x[x].index
```

```python
rating_users = x[x].index
```

```python
filtered_rating = df1[df1['userID'].isin(rating_users)]
```

```python
y = filtered_rating.groupby('title').count()['rating'] >= 2
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = df1[df1['title'].isin(famous_books)]
```

```python
final_ratings
```

```python
#dropping duplicates if any

final_ratings = final_ratings.drop_duplicates()
```

```python
# This is our final books for recommendation.
final_ratings
```

```python
pt = final_ratings.pivot_table(index='title', columns='userID', values='rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

- Now we have 104 books to recommend on basis of 113 users ratings

# Model Building

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:4]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df1[df1['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('White Oleander : A Novel')
```

```python
recommend('A Little Princess')
```

```python
recommend('Watership Down')
```
