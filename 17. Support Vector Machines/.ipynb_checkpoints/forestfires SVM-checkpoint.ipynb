# Problem Statement:

classify the Size_Categorie using SVM

month	month of the year: 'jan' to 'dec'
day	day of the week: 'mon' to 'sun'
FFMC	FFMC index from the FWI system: 18.7 to 96.20
DMC	DMC index from the FWI system: 1.1 to 291.3
DC	DC index from the FWI system: 7.9 to 860.6
ISI	ISI index from the FWI system: 0.0 to 56.10
temp	temperature in Celsius degrees: 2.2 to 33.30
RH	relative humidity in %: 15.0 to 100
wind	wind speed in km/h: 0.40 to 9.40
rain	outside rain in mm/m2 : 0.0 to 6.4
Size_Categorie 	the burned area of the forest ( Small , Large)

# Data Preprocessing

### Identifying Duplicates

```python
df.duplicated().sum()
```

### Identifying Outliers

```python
plt.figure(figsize=(8,8),dpi=200)
sns.boxplot(data=df)
plt.xticks(rotation=90);
```

### Remove unnecessary columns

```python
df = df.drop(['dayfri', 'daymon', 'daysat', 'daysun', 'daythu',
       'daytue', 'daywed', 'monthapr', 'monthaug', 'monthdec', 'monthfeb',
       'monthjan', 'monthjul', 'monthjun', 'monthmar', 'monthmay', 'monthnov',
       'monthoct', 'monthsep'], axis=1)
```

```python
df.head()
```

# Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['month'] = label_encoder.fit_transform(df['month']) 
df['day'] = label_encoder.fit_transform(df['day'])
df['size_category'] = label_encoder.fit_transform(df['size_category'])
```

```python
df.head()
```

# Data Visualization

```python
sns.pairplot(data=df, hue='size_category');
```

```python
plt.figure(figsize = (14, 6));
sns.heatmap(df.corr(), cmap='magma', annot=True, fmt=".3f")
plt.show()
```

# Data Splitting

```python
X = df.drop('size_category',axis=1)
y = df['size_category']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

# Model Selection

```python
from sklearn.svm import SVC
```

```python
base_model = SVC()
```

# Train Model

```python
base_model.fit(scaled_X_train,y_train)
```

# Model Evaluation

```python
y_pred = base_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
accuracy_score(y_test, y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=base_model.classes_).plot();
```

```python
print(classification_report(y_test, y_pred))
```

# Cross Validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC(class_weight='balanced')
```

```python
param_grid = {'C':[0.001,0.01,0.1,0.5,1],
             'kernel':['linear','rbf','poly'],
              'gamma':['scale','auto']}
```

```python
grid_model = GridSearchCV(svm,param_grid,verbose=3)
```

```python
grid_model.fit(scaled_X_train,y_train)
```

```python
grid_model.best_params_
```

# Model Evaluation

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=grid_model.classes_).plot();
```

```python
print(classification_report(y_test, y_pred))
```
